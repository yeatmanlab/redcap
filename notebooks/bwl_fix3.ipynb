```python
# Just a small script to prepare current data in the adu_block csv into REDCap
import pandas as pd
import numpy as np
import os
import glob
```

```python
# Import the dataset
bwl_data = pd.read_csv('/Users/douglasstrodtman/Downloads/adu_block_scores.csv', index_col=['nlr_id'])
```

```python
# import ids from repo
id_data = pd.read_csv('/Users/douglasstrodtman/Downloads/RDRPRepository_DATA_2017-04-11_1617.csv', index_col='nlr_id')
```

```python
# join data
bwl_wid = bwl_data.combine_first(id_data)
```

```python
# select only subs with block scores
bwl_wid = bwl_wid.loc[pd.notnull(bwl_wid.bwl_correct)]
```

```python
bwl_wid.loc[bwl_wid['visit']==1, 'redcap_event_name'] = 'visit_1_arm_1'
bwl_wid.loc[bwl_wid['visit']==2, 'redcap_event_name'] = 'visit_2_arm_1'
bwl_wid.loc[bwl_wid['visit']==3, 'redcap_event_name'] = 'visit_3_arm_1'
```

```python
bwl_wid
```

```python
# add sub indices
bwl_wid = bwl_wid.set_index(['visit', 'bwl_form'], append=True)
```

```python
list1 = bwl_wid.loc[bwl_wid['bwl_list'] == 1]
```

```python
# concat visits into single row

list1 = bwl_wid.loc[bwl_wid['bwl_list'] == 1]
list1_fixed = list1[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list1_fixed.columns = ['bwl_correct1', 'bwl_time1', 'record_id', 'redcap_event_name']
list2 = bwl_wid.loc[bwl_wid['bwl_list'] == 2]
list2_fixed = list2[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list2_fixed.columns = ['bwl_correct2', 'bwl_time2', 'record_id', 'redcap_event_name']
list3 = bwl_wid.loc[bwl_wid['bwl_list'] == 3]
list3_fixed = list3[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list3_fixed.columns = ['bwl_correct3', 'bwl_time3', 'record_id', 'redcap_event_name']
list4 = bwl_wid.loc[bwl_wid['bwl_list'] == 4]
list4_fixed = list4[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list4_fixed.columns = ['bwl_correct4', 'bwl_time4', 'record_id', 'redcap_event_name']
list5 = bwl_wid.loc[bwl_wid['bwl_list'] == 5]
list5_fixed = list5[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list5_fixed.columns = ['bwl_correct5', 'bwl_time5', 'record_id', 'redcap_event_name']
list6 = bwl_wid.loc[bwl_wid['bwl_list'] == 6]
list6_fixed = list6[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list6_fixed.columns = ['bwl_correct6', 'bwl_time6', 'record_id', 'redcap_event_name']
list7 = bwl_wid.loc[bwl_wid['bwl_list'] == 7]
list7_fixed = list7[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list7_fixed.columns = ['bwl_correct7', 'bwl_time7', 'record_id', 'redcap_event_name']
list8 = bwl_wid.loc[bwl_wid['bwl_list'] == 8]
list8_fixed = list8[['bwl_correct', 'bwl_time', 'record_id', 'redcap_event_name']]
list8_fixed.columns = ['bwl_correct8', 'bwl_time8', 'record_id', 'redcap_event_name']
```

```python
# finish combining
bwl_fixed = list1_fixed.combine_first(list2_fixed)
bwl_fixed = bwl_fixed.combine_first(list3_fixed)
bwl_fixed = bwl_fixed.combine_first(list4_fixed)
bwl_fixed = bwl_fixed.combine_first(list5_fixed)
bwl_fixed = bwl_fixed.combine_first(list6_fixed)
bwl_fixed = bwl_fixed.combine_first(list7_fixed)
bwl_fixed = bwl_fixed.combine_first(list8_fixed)
```

```python
bwl_fixed = bwl_fixed.assign(bwl_run = 3)
```

```python
bwl_fixed.to_csv('/Users/douglasstrodtman/Downloads/record_right.csv')
```
